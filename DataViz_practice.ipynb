## Histogram and distribution using matplot lib and oop method

```python
# matplot lib method
# salary
plt.figure(figsize=(5,3))
plt.hist(HRData['Salary'], bins=20, color='Blue', edgecolor='white')
plt.legend(labels=['salary'])
plt.title('Dirtibution of salary', fontsize = 16)
plt.xlabel('salary', fontsize = 13)
plt.ylabel('Frequency', fontsize = 13)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3) # alpha is representation of tranperency, 1 being opque and 0 being transperent
plt.show()

# EmpSatisfaction
plt.figure(figsize=(5,3))
plt.hist(HRData['EmpSatisfaction'], bins=20, color='black', edgecolor='white')
plt.legend(labels=['EmpSatisfaction'])
plt.title('Dirtibution of EmpSatisfaction', fontsize = 16)
plt.xlabel('EmpSatisfaction', fontsize = 13)
plt.ylabel('Frequency', fontsize = 13)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3) # alpha is representation of tranperency, 1 being opque and 0 being transperent
plt.show()

# EngagementSurvey
plt.figure(figsize=(5,3))
plt.hist(HRData['EngagementSurvey'], bins=20, color='skyblue', edgecolor='white')
plt.legend(labels=['EngagementSurvey'])
plt.title('Dirtibution of EngagementSurvey', fontsize = 16)
plt.xlabel('EngagementSurvey', fontsize = 13)
plt.ylabel('Frequency', fontsize = 13)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3) # alpha is representation of tranperency, 1 being opque and 0 being transperent
plt.show()
```

```python
# oop method
# Set up subplots with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot 1: Distribution of Salary
axes[0].hist(HRData['Salary'], bins=20, color='blue', edgecolor='white')
axes[0].set_title('Distribution of Salary', fontsize=16)
axes[0].set_xlabel('Salary', fontsize=13)
axes[0].set_ylabel('Frequency', fontsize=13)
axes[0].tick_params(axis='x', labelsize=10, rotation=45)
axes[0].tick_params(axis='y', labelsize=10)
axes[0].grid(alpha=0.3)

# Plot 2: Distribution of EmpSatisfaction
axes[1].hist(HRData['EmpSatisfaction'], bins=20, color='black', edgecolor='white')
axes[1].set_title('Distribution of EmpSatisfaction', fontsize=16)
axes[1].set_xlabel('EmpSatisfaction', fontsize=13)
axes[1].set_ylabel('Frequency', fontsize=13)
axes[1].tick_params(axis='x', labelsize=10, rotation=45)
axes[1].tick_params(axis='y', labelsize=10)
axes[1].grid(alpha=0.3)

# Plot 3: Distribution of EngagementSurvey
axes[2].hist(HRData['EngagementSurvey'], bins=20, color='skyblue', edgecolor='white')
axes[2].set_title('Distribution of EngagementSurvey', fontsize=16)
axes[2].set_xlabel('EngagementSurvey', fontsize=13)
axes[2].set_ylabel('Frequency', fontsize=13)
axes[2].tick_params(axis='x', labelsize=10, rotation=45)
axes[2].tick_params(axis='y', labelsize=10)
axes[2].grid(alpha=0.3)

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

## Box Plots for Categorical Variables

```python
sns.set(style='whitegrid')
plt.figure(figsize=(7,3))
sns.boxplot(x='MarriedID', y='Salary', data=HRData, palette='Set3')
plt.title('Number of marriages', fontsize = 16)
plt.xlabel('MarriedID', fontsize = 13)
plt.ylabel('Salary', fontsize = 13)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Correlation Heatmap

```python
# corrilation matrix
corr_matrix = HRData.corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(9, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.2)

# Customize the plot
plt.title('Correlation Heatmap of Numerical Variables', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Display the plot
plt.show()
```

##  pair plot

```python
selected_vars = ['Salary', 'EngagementSurvey', 'EmpSatisfaction', 'SpecialProjectsCount']

# Set up the pair plot
sns.set(style="ticks") # ticks = mark the ticks on the graph
sns.pairplot(HRData[selected_vars], height=2, markers='s', diag_kind='kde', aspect=1.1) 
# markers: s = squear, o = circle, d = dimond, +, *, x
# diag_kind: kde =  Kernel Density Estimate, hist and None- None will not plot anytghing on the graph

# Customize the plot
plt.suptitle('Pair Plot of Selected Numerical Variables', x=0.5, y=1.02, fontsize=16)
plt.show()
```

## bar plot for Categorical Variables

```python
plt.figure(figsize=(7, 3))
ax = sns.countplot(x='DeptID', data=HRData, palette='viridis')

# this code is for the putting the value over the bar for y parameter
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 2), textcoords='offset points', fontsize=10)
    
plt.title('Number of Employees in Each Department', fontsize=16)
plt.xlabel('Department ID', fontsize=14)
plt.ylabel('Number of Employees', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

## Scatter Plot

## Pie Charts

```python
marital_status_counts = HRData['MarriedID'].value_counts()

# Map values to labels
marital_status_labels = {0: 'Single', 1: 'Married'}
marital_status_counts.index = marital_status_counts.index.map(marital_status_labels)

plt.figure(figsize=(8, 8))
pie, texts, autotexts = plt.pie(marital_status_counts, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=0)

# Add legend annotations
legend_labels = ['Single', 'Married']
legend_colors = ['skyblue', 'lightcoral']
legend_patches = [plt.Line2D([0], [0], marker='*', color='k', label=label, markersize=15, markerfacecolor=color) for label, color in zip(legend_labels, legend_colors)]
plt.legend(handles=legend_patches, loc='upper right')

# Add text annotations
for text, autotext, label in zip(texts, autotexts, marital_status_counts.index):
    text.set_text(f'{label}: {text.get_text()}')

plt.title('Distribution of Employees by Marital Status', fontsize=16)

plt.show()
```
